```python
import base64
import zlib
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('map1.tmx')
root = tree.getroot()
```

```python
def get_layers(tree):
    for layer in root.findall('layer'):
        yield layer
```

```python
layers = get_layers(root)
```

```python
layer = next(layers)
```
